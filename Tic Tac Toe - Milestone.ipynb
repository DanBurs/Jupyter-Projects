```python
from IPython.display import clear_output
import random

def display_game1(moves):
    clear_output()   
    
    print("Here is the current board:  ")
    row1 = [moves[7]+'|'+moves[8]+'|'+moves[9]]
    row2 = [moves[4]+'|'+moves[5]+'|'+moves[6]]
    row3 = [moves[1]+'|'+moves[2]+'|'+moves[3]]
    board = [row1,row2,row3]
    
    for p in board:
        for place in p:
            print(place)

def marker_choice():
    marker = ''
    while marker != 'X' and marker !='O':
        marker = input("Player 1 - Choose 'X' or 'O': ").upper()
        
    player1 = marker
    if player1 == 'X':
        player2 = 'O'
    else: 
        player2 = 'X'
        
    return (player1, player2)

def place_marker(board, marker, position):
    board[position] = marker
    
def win_check(board, mark):
    if mark in board:
        if board[7]==board[8]==board[9]==mark:
            return True
            game_on = False
        elif board[4]==board[5]==board[6]==mark:
            return True
            game_on = False
        elif board[1]==board[2]==board[3]==mark:
            return True
            game_on = False
        elif board[7]==board[4]==board[1]==mark:
            return True
            game_on = False
        elif board[8]==board[5]==board[2]==mark:
            return True
            game_on = False
        elif board[9]==board[6]==board[3]==mark:
            return
            game_on = False
        elif board[7]==board[5]==board[3]==mark:
            return True
            game_on = False
        elif board[1]==board[5]==board[9]==mark:
            return True
            game_on = False
        else:
            return False
        
def choose_first():
    if random.randint(1,2)==1:
        print('Player 1 Goes First\n\n')
        return int(1)
    else:
        print('Player 2 Goes First\n\n')
        return int(2)
    
def space_check(board, position):
    return board[position] == '_'

def full_board(board):
    for i in range(1,10):
        if space_check(board,i):
            return False
    return True 

def user_choice(board):
    #variables
    #Initial choice
    num = "No"
    choice=0
    acceptable_range=range(0,10)
    within_range = False
    
    #Two conditions to check
    #digit or within range
    while space_check(board, choice) == False or within_range == False:
    #digit check
        choice = int(input("Please enter a number (0-9): "))
        num = str(choice)
        if num.isdigit() == False:
            print("Sorry, that's not a digit ")
            break
    #range check
        if num.isdigit() == True: 
            if choice in acceptable_range:
                within_range = True  
            else:
                print("Sorry, that's not within the range")
                within_range = False
    return choice
    
def gameon():
    
    choice1 = 'wrong'
    
    while choice1 not in ['Y','N']:
        
        choice1 = input ("Keep Playing? (Y or N): ").upper()        
        
        if choice1 not in ['Y','N']:
            print ("Sorry, please choose Y or N")
        elif choice1 == "Y":
            return True
        else:
            return False
            
        
# Now the Game

print('Welcome to Tic Tac Toe!\n\n')

playing = True

while playing:
    #Set the game up here
    board = ["_"]*11
    player1_mark , player2_mark = marker_choice() 
    turn = choose_first()
    
    play = input('Are you ready?!? Enter: y or n')
    
    if play.upper()[0] == 'Y':
        game_on = True
    else:
        game_on = False

    while game_on:
        
        if turn == 1:
            #Player 1 Turn
            display_game1(board)
            place_marker(board, player1_mark,user_choice(board))

            if win_check(board, player1_mark) == True:
                display_game1(board)
                print('PLAYER 1 WINS!')
                game_on = False

            else:
                if full_board(board):
                    display_game1(board)
                    print("This is a draw!")
                    break
                else:
                    turn = 2 
                

        else:
            # Player2's turn.
            display_game1(board)
            place_marker(board, player2_mark,user_choice(board))
            
            if win_check(board, player2_mark) == True:
                display_game1(board)
                print('PLAYER 2 WINS!')
                game_on = False

            else:
                if full_board(board):
                    display_game1(board)
                    print("This is a draw!")
                    break
                else:
                    turn = 1
                
    playing = gameon()
        
    if not playing:
        clear_output()
        break 
    else:
        clear_output()
        continue
    
print("\n\nThanks for Playing!\n\n")
```
